```python
r = requests.get('https://g1.globo.com/')
r.encoding = 'utf-8' #r.apparent_encoding
soup = BeautifulSoup(r.text, "html.parser")
```

```python
r.apparent_encoding
```

```python
lista = soup.find_all('a', class_='feed-post-link gui-color-primary gui-color-hover')
```

```python
lista
```

```python
import pandas as pd
ldf = []
for l in lista:
    ldf.append([l.text, l.get('href')])

pd.DataFrame(ldf, columns=['titulo', 'link'])
```

```python
url = 'https://www.barrashopping.com.br/lojas'
r = requests.get(url)
r.encoding = r.apparent_encoding
content = r.text
```

```python
#content
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
l = soup.find_all('li', class_='establishment')
```

```python
l[0]
```

```python
ldf = []
for i in l:
    s2 = BeautifulSoup(str(i), "html.parser")
    
    ldf.append([s2.find_all('p', class_='nameRegex hide')[0].text,
                s2.find_all('a', class_='details')[0].text])
```

```python
df = pd.DataFrame(ldf, columns=['titulo', 'tel'])
```

```python
df
```
