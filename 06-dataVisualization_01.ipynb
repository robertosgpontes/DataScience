```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_json(
    path_or_buf='https://api.bcb.gov.br/dados/serie/bcdata.sgs.4390/dados?formato=json')
df['data'] = pd.to_datetime(df['data'])
```

### Plot usando funções do dataframe:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html

```python
df.plot(
    x='data', 
    y ='valor', 
    kind='line')
```

```python
df.plot(
    x='data', 
    y ='valor', 
    kind='line',
    grid=True,
    rot=90)
```

```python
df[df.data > '01-01-1996'].plot(
    x='data', 
    y ='valor', 
    kind='line',
    grid=False,
    figsize=(20, 16))
```

```python
df2 = pd.DataFrame([[i, i, i**2, i**3] for i in range(0, 10000)], columns=['x', 'y=x', 'y=x**2', 'y=x**3'])
```

```python
df2.plot(
    x='x'
)
```

```python
df2.plot(
    x='x',
    logy=True
)
```

### Usando Matplotlib
https://matplotlib.org/1.4.1/api/pyplot_api.html#matplotlib.pyplot.plot

```python
plt.plot(
    df.data,
    df.valor
)
plt.show()
```

```python
plt.plot(
    df.data,
    df.valor,
    '-'
)
plt.title('Selic por Mês')
plt.xlabel('Mês/Ano')
plt.ylabel('Taxa Selic')
plt.show()
```

```python
fig = plt.figure(figsize=(20, 8))

ax = fig.add_subplot(121)
df.plot(
    x='data', 
    y='valor', 
    kind='line',
    ax=ax)
ax.set_title('Selic por Mês')
ax.set_xlabel('Mês/Ano')
ax.set_ylabel('Taxa Selic')

ax2 = fig.add_subplot(122)
df[df.data > '01-01-1996'].plot(
    x='data', 
    y ='valor', 
    kind='line',
    grid=False,
    ax=ax2)
```

```python
fig = plt.figure(figsize=(20, 16))

ax = fig.add_subplot(221)
df.plot(
    x='data', 
    y='valor', 
    kind='line',
    ax=ax)
ax.set_title('Selic por Mês')
ax.set_xlabel('Mês/Ano')
ax.set_ylabel('Taxa Selic')

ax2 = fig.add_subplot(222)
df[df.data > '01-01-1996'].plot(
    x='data', 
    y ='valor', 
    kind='line',
    grid=True,
    ax=ax2)

ax2 = fig.add_subplot(223)
df2.plot(
    x='x',
    ax=ax2
)

ax2 = fig.add_subplot(224)
df2.plot(
    x='x',
    logy=True,
    ax=ax2
)
```
