```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xdata = np.arange(1, 11, 0.4)
ydata = np.random.randn(len(xdata))

xdata2 = np.arange(0, 11, 0.2)
ydata2 = np.sort(10 * np.exp(-xdata2 / 4) + np.random.randn(len(xdata2)) / 3)
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.plot(xdata, ydata, label='Plotted data')
ax.plot(xdata2, ydata2, label='Transform data')

ax.set_xlabel('X [m]')
ax.legend()

plt.show() # Fora do Jupyter
```

```python
fig, ax = plt.subplots(constrained_layout=True)
ax.plot(xdata, ydata, label='Plotted data')

ax.plot(xdata2, ydata2, label='Transform data')
ax.set_xlabel('X [m]')
ax.legend()

secax = ax.secondary_xaxis('top')
secax.set_xlabel('$X_{other}$')

plt.show()
```

```python
s = np.around(np.random.normal(170, 25, 1000000),1)
```

```python
print(s)
```

```python
count, bins, ignored = plt.hist(s, 100)
plt.show()
```

```python
count, bins, ignored = plt.hist(s, 100, density=True)
plt.show()
```

```python
sigma = s.std()
mu = s.mean()
```

```python
sigma
```

```python
count, bins, ignored = plt.hist(s, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.title('Altura da População')
plt.xlabel('BINS')
plt.ylabel('PDF')
plt.show()
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 15]
```

```python
explode = (0.05, 0.05, 0.05, 0.05) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
